```python
import pandas as pd
import numpy as np

persontype = np.dtype({
    'names':['name','chinese','math','english'],
    'formats':['S32','i','i','i']
})
persons = np.array([
    ("Zhang Fei",85,100,90),
    ("Guan Yu",85,96,85),
    ("Zhao Yun",88,92,96.5),
    ("Dian shu",99,85,100)
],dtype=persontype)

df1 = pd.DataFrame(persons)

result1 = df1.describe()
print(result1)
max = result1.loc['max',['chinese','math','english']]
print(max)
min = result1.loc['min',['chinese','math','english']]
print(min)
mean = result1.loc['mean',['chinese','math','english']]
print(mean)
std = result1.loc['std',['chinese','math','english']]
print(std)

```

```python
import pandas as pd

df = pd.read_csv('./car_complain.csv')

df = df.drop('problem',axis=1).join(df.problem.str.get_dummies(','))

print(df)
```

```python
def f(x):
    x = x.replace('一汽-大众','一汽大众')
    return x

df['brand'] = df['brand'].apply(f)

result = df.groupby(['brand'])['id'].agg(['count'])
result
```

```python
result1 = df.groupby(['car_model'])['id'].agg(['count'])
result1.sort_values('count',ascending=False)
```

```python
import numpy as np
result = df.groupby(['brand','car_model'])['id'].agg(['count'])
result
```

```python
tags = df.columns[7:]
result2 = df.groupby(['brand'])[tags].agg(['sum'])
result2
```

```python
result2.reset_index(inplace=True)
result2
```

```python
result2.to_csv("./result.csv")
```
