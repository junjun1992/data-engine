```python
X = data.iloc[:,[1,4]].values
print(type(X))
print(X.shape)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores =[]
range_values = np.arange(2, 10)
for i in range_values:
    kmeans = KMeans(init='k-means++',n_clusters= i, max_iter = 300, n_init = 10, random_state=0)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_,metric='euclidean',sample_size=len(X))
    scores.append(score)
    
plt.figure()
plt.bar(range_values,scores,width=0.6,color='b',align='center')
plt.title('silhoutette score')
plt.show()
```

```python
plt.plot(range_values,scores)
plt.title('silhoutette score')
plt.xlabel('Cluster')
plt.ylabel('score')
plt.show()
```

```python
#k=2
km = KMeans(init = 'k-means++',n_clusters = 2,max_iter = 300,n_init = 10,random_state = 0 , n_jobs=-1)
y_means = km.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],marker='o',facecolors = 'b',edgecolors='k',s =30)
centroids = km.cluster_centers
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],s=100,c='pink',label = 'm')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],s=100,c='yellow',label = 'g')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],s=100,c='cyan',label = 't')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],s=100,c='magenta',label = 's')
plt.scatter(X[y_means == 4,0],X[y_means == 4,1],s=100,c='orange',label = 'c')
plt.scatter(centroids[:,0],centroids[:,1],marker ='*',s=260,linewidths=3,color='black',label='a')
x_min,x_max = min(X[:,0])-1,max(X[:,0])+1
y_min,y_max = min(X[:,1])-1,max(X[:,1])+1
plt.title('K-Means clustering')
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(X,quantile=0.1)
ms = MeanShift(bandwidth=bandwidth).fit(X)

plt.scatter(X[:,0],X[:,1],marker='o',facecolors = 'b',edgecolors='k',s =30)
centroids = ms.cluster_centers
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],s=100,c='pink',label = 'm')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],s=100,c='yellow',label = 'g')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],s=100,c='cyan',label = 't')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],s=100,c='magenta',label = 's')
plt.scatter(X[y_means == 4,0],X[y_means == 4,1],s=100,c='orange',label = 'c')
plt.scatter(centroids[:,0],centroids[:,1],marker ='*',s=260,linewidths=3,color='black',label='a')
x_min,x_max = min(X[:,0])-1,max(X[:,0])+1
y_min,y_max = min(X[:,1])-1,max(X[:,1])+1
plt.title('Mean shift clustering')
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
